```python
# 일별 수익률 생성 : 전날주가 / 오늘주가
# price_aapl에 새로운 파생변수 생성하여 일별 수익률(daily_rtn)대입

price_aapl['전날주가'] = price_aapl['Adj Close'].shift(1)
```

```python
price_aapl
```

```python
price_aapl['daily_rtn'] = price_aapl['Adj Close'] / price_aapl['전날주가'] 
```

```python
price_df = price_aapl.tail(1000)
```

```python
price_df.iloc[0,2]
```

```python
# 함수(buyandhold)를 생성 -> 매개변수 df col start end

def buyandhold(df, col, start, end):
    # df에 결측치나 무한대 값들을 모두 제거
    df = df.loc[~(df.isin([np.nan, np.inf,-np.inf]).any(axis=1))]
    # df에 있는 date컬럼을 시계열 데이터로 변경
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
    # col 매개변수를 이용하여 df에 해당하는 col컬럼만 제외하고 모두 제거
    df = df[[col]]
    # start,end 데이터를 시계열 데이터로 변경
    start = datetime.strptime(start, '%Y%m').isoformat()
    end = datetime.strptime(end, '%Y%m').isoformat()
    
    # daily_rtn 파생변수를 생성하여 일일 수익률 계산
    df['daily_rtn'] = df[col].pct_change()

    # start end를 기준으로 df필터링
    df = df.loc[start:end]

    # daily_rtn 파생변수를 생성하여 누적 수익률 계산
    df['rtn'] =(1+ df['daily_rtn']).cumprod()

    # 누적수익률의 마지막 데이터
    result = df['rtn'][-1]
    # 데이터 프레임 리턴
    return df, result

    
```

```python
df = pd.read_csv('../csv/csv2/csv3/AAPL.csv')
```

```python
buyandhold(df,'Adj Close','201805','201905')
```
